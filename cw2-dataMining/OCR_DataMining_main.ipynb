Build models based on the corpus we have now.

```python
# clean it to make it can be used in the vectorization
modelCorpus=[]
for i in corpus["contents"]:
    text_temp=" ".join([item for item in i])
    modelCorpus.append(text_temp)
pickle.dump(modelCorpus, open("modelCorpus", "wb"))
# modelCorpus=pickle.load(open("modelCorpus","rb"))
```

```python
modelCorpus=pickle.load(open("modelCorpus","rb"))
```

```python
# TFIDF 
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer()
transformer=TfidfTransformer()
tfidf=transformer.fit_transform(vectorizer.fit_transform(modelCorpus))
vocabulary=vectorizer.get_feature_names()
weight=tfidf.toarray() 

# Note:
# weight is the tf-idf matrix, a[i][j] is the weight of j word in i document
# Each row can be seen as a vector for each book
```

```python
# show the feature size of the TF-IDF vectors
weight.shape
```

```python
# Doc2vec
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import logging
```

```python
# Only need to uncomment this when I need to re-train this model.

max_epochs = 100
alpha = 0.025

documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(modelCorpus)]
model = Doc2Vec(documents, vector_size=100,alpha=alpha, min_alpha=0.00025, window=5, min_count=10)
# model.build_vocab(documents)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(documents,total_examples=model.corpus_count,epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
```

**check the distance between different documents**

**Several Distance Measurements**:
$$ cosine\space similarity = \frac{X\cdot Y}{|X|\cdot |Y|}$$
$$ cosine\space distance = 1- cosine \space similarity$$
$$ euclidean\space distance = (X-Y)^2$$

Cosine distance mainly measure the difference of the direction of vector

Euclidean distance mainly measure the distance between the data 

**And then**:

Draw the plot to show the visualization

First way calculate the distance of different vectors:

```python
from scipy import spatial
import numpy as np
```

```python
# Calculate the distance based on the Doc2vec
distance_matrix_doc2vec=np.ndarray((24,24))
for i in range(24):
    for j in range(24):
        distance_matrix_doc2vec[i][j] = spatial.distance.cosine(model.docvecs[i], model.docvecs[j])
```

```python
# Calculate the distance based on the tfidf
distance_matrix_tfidf=np.ndarray((24,24))
for i in range(24):
    for j in range(24):
        distance_matrix_tfidf[i][j] = spatial.distance.cosine(weight[i], weight[j])
```

**elbow method for TF-IDF**

```python
# based on the weight(TF-IDF) 
distortions = []
K = range(1, 20)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(weight)
#     kmeanModel.fit(weight)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of squared distances to cluster center')
plt.title('The Elbow Method for TF-IDF vectors')
plt.show()
```

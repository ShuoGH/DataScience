```python
xv=np.linspace(3.85,4.15,1000) # appropriate x-axis limits
#this is used to show the gaussian fig of the uniform distribution
def gaussianvec(x, slen, p):
    mu = p # mean is the same as that of the Bernoulli generating dist
    sig = np.sqrt(4/(12*slen)) # s.d.    standard variance
    return np.asarray(np.exp(-(x-mu)**2/(2*sig**2))/(np.sqrt(2*np.pi)*sig))

fig, ax = plt.subplots(nrows=2,ncols=3,figsize=(15,8))
#this is used to plot the figures.  2*3   six figures

for i, s in enumerate(sig):    #sig [10,20,30]   i [0,1,2]
    #np.mean()   how to use this function?
#     ax[0][0].hist(np.mean(ranblock0[:s**2],axis=0),alpha=0.3,normed=True)  
    ax[0][0].hist(np.mean(ranblock0[:s**2],axis=0),alpha=0.3,density=True)  #'normed'has been deprecated.
    ax[0][0].hist(np.mean(ranblock0[:],axis=0),alpha=0.3,density=True)
    ax[0][1].plot(xv,gaussianvec(xv, s**2,4))
    ax[0][2].hist(np.mean(ranblock0[:s**2],axis=0),alpha=0.3,density=True) #the density is used to do whether it's normalized 
    ax[0][2].plot(xv,gaussianvec(xv, s**2,4),alpha=0.9)
    ax[1][0].hist(np.mean(ranblock1[:s**2],axis=0),alpha=0.3,density=True)
    ax[1][1].plot(xv,gaussianvec(xv, s**2,4))
    ax[1][2].hist(np.mean(ranblock1[:s**2],axis=0),alpha=0.3,density=True)
    ax[1][2].plot(xv,gaussianvec(xv, s**2,4),alpha=0.9)
    ax[1][0].set_ylabel(r'bin frequency in $10^5$ sequences')
    ax[0][0].set_ylabel(r'bin frequency in $10^2$ sequences')
    
#ax[0][0].hist(np.mean(ranblock0[:],axis=0),density=True)    
#np.mean()     axis the calculate should do along the flat one. so this mean that the experiment do 100 or 100000times
```

### CLT: the coin tossing example 
it's a binary....what

```python
nreps = [10**2, 10**5]
sig = [10,20,30]
seqlen = [i**2 for i in sig]
ranblock0=np.random.binomial(1,.5,(seqlen[-1],nreps[0])) # generating longest sequence 
ranblock1=np.random.binomial(1,.5,(seqlen[-1],nreps[1])) # subsequences used below

xv=np.linspace(0.35,0.65,1000)
def gaussianvec(x, slen, p):
    mu = p # mean is the same as that of the Bernoulli generating dist (binomial)
    sig = np.sqrt(p*(1-p)/slen) # s.d. of binomial = sqrt(N*p*q)/N
    return np.asarray(np.exp(-(x-mu)**2/(2*sig**2))/(np.sqrt(2*np.pi)*sig))
fig, ax = plt.subplots(nrows=2,ncols=3,figsize=(15,8))
for i, s in enumerate(sig):
    ax[0][0].hist(np.mean(ranblock0[:s**2],axis=0),alpha=0.3,density=True)
    ax[0][1].plot(xv,gaussianvec(xv, s**2,0.5))
    ax[0][2].hist(np.mean(ranblock0[:s**2],axis=0),alpha=0.2,density=True)
    ax[0][2].plot(xv,gaussianvec(xv, s**2,0.5),alpha=0.4)
    ax[1][0].hist(np.mean(ranblock1[:s**2],axis=0),alpha=0.3,density=True)
    ax[1][1].plot(xv,gaussianvec(xv, s**2,0.5))
    ax[1][2].hist(np.mean(ranblock1[:s**2],axis=0),alpha=0.2,density=True)
    ax[1][2].plot(xv,gaussianvec(xv, s**2,0.5),alpha=0.4)

    ax[1][0].set_ylabel(r'bin frequency in $10^5$ sequences')
    ax[0][0].set_ylabel(r'bin frequency in $10^2$ sequences')

#didn't review wholely
```

### what proportion of the earth's surface is water?
(from "rethinking statistics" by Richard McElreath)
Let the unknown proprtion of water be denoted  θ theta

```python
def posterior_grid_approx(grid_points,success,tosses):
    #define grid: introduce points theta 
    #posteriot probability, p(theta) is returned
    p_grid=np.linspace(0,1,grid_points)   # start stop num
    
    #define prior
    prior=np.repeat(1,grid_points)/grid_points
    
    #compute likelihood at each point in the grid
    likelihood=stats.binom.pmf(success,tosses,p_grid)
    
    #compute product of likelihood and prior
    unstd_posterior=likelihood*prior
    
    #standardize the posterior, so it sums to 1
    posterior=unstd_posterior/unstd_posterior.sum()
    return p_grid,posterior
```

```python
#try the stats.binom.pmf   pobability mass function
#x denotes the probability, n denotes the whole times, k denotes the certain times
def bin(x,n,k):
    return math.factorial(n)/(math.factorial(k)*math.factorial(n-k))*(x**k)*(1-x)**(n-k)
#what's factorial() function 阶乘
point_grid=np.linspace(0,1,6) #an array of 6 points
# for i in point_grid:
#     print(i)
print("check if we understand what stats.binom.pmd returns. are the following two lines equal?")

print(bin(point_grid,10,7))      
#input the sequence of the possbabilites  0 0.2 0.4 0.6 0.8 1 
print(stats.binom.pmf(7,10,point_grid))
stats.binom.pmf(1,3,0.2)  #should be 0.384 
#yep they are equal 
```

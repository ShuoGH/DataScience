## Introduction to PCA

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as spio
import seaborn as sns; sns.set()
from mpl_toolkits.mplot3d import Axes3D
```

First some snythetic data (5000 points) is generated from a tri-variate gaussian distribution

```python
rng=np.random.RandomState(1)
Xmean=[2,4,-8]
Xcov=[[8.5,-3.5,2.1],[-3.5,12.,4.5],[2.1,4.5,3.5]]
print(np.linalg.eigh(Xcov))
npts=10000
X = np.random.multivariate_normal(Xmean,Xcov,npts)
print(X,X.shape)
fig=plt.figure(figsize=(18,6))
ax0=fig.add_subplot(121,projection='3d')
ax0.scatter(X[:,0],X[:,1],X[0:,2],alpha=0.05)
# 
# doc: np.linalg.eigh  ax0.scatter
```

```python
Xsmean=np.mean(X,axis=0)
Xdev=X-Xsmean
# print(Xdev.shape) #shape 10000,3
# generalise the mean to zero.
S=np.dot(Xdev.T,Xdev)/(npts-1)
#  what's this for....***********

el,ev=np.linalg.eigh(S)
print(el,ev)
# el1,ev1=np.linalg.eig(S)
# print(el1,ev1)    
# 两个计算结果不太一样。特征值一样，特征向量不同。
print(el/el.sum())
# 特征值归一化？？？
# print(ev[1]) #TEST the [] index for row.
```

```python
fig, ax = plt.subplots(ncols=3, nrows=1, figsize=(16,5))
print(Xcov)
print(np.cov(Xdev.T))
# calculate the covariance matrix of the Xdev. 
# need to transpose the matrix to 3*10000 first, the cov() will calculate out the 3*3 cov matrix

for i in range(3):
    print(np.cov(np.dot(Xdev,ev[2-i])))
#   将Xdev和特征向量的第2-i行点乘. 10000*3 矩阵和 3*1 的向量相乘。会变成对应的特征值吧？ 然后求cov 实际上是自己的方差。
    ax[i].hist(np.dot(Xdev,ev[2-i]))
#  都是正态分布。
    
```

```python
Xproj = np.dot(Xdev,ev)
print(Xproj,Xproj.shape)
```

```python
fig, ax = plt.subplots(ncols=5, nrows=1, figsize=(16,3))
# draw five figures to the subplots

print(np.cov(Xproj.T)) 
# output the cov matrix 
print(np.mean(Xproj,axis=0))
# calculate the mean of the Xproj along the row(calculate the mean for each column)

for i in range(3):
    ax[i].hist(Xproj[:,2-i],density=True)
    ax[3].hist(Xproj[:,2-i],alpha=0.3,density=True)
    ax[4].hist(Xproj[:,2-i],alpha=0.3)
```

you should notice that the new co-ordinates of the data have different variances. To facilitate the comparison, draw two another histograms, the rightmost figure is not normalised. The sencond rightmost one is normalised.

```python
fig, ax = plt.subplots(nrows=2,ncols=3,figsize=(15,6))
# draw six plots 

for i in range(2):
    for j in range(3):
        ax[i][j].set_aspect(1,'datalim')
        
ax[0][0].scatter(X[:,2],X[:,1],alpha=0.01)
ax[0][1].scatter(X[:,2],X[:,0],alpha=0.01)
ax[0][2].scatter(X[:,1],X[:,0],alpha=0.01)
# X is the data generated from the gaussian distribution 

ax[1][0].scatter(Xproj[:,2],Xproj[:,1],alpha=0.01,c='g')
ax[1][1].scatter(Xproj[:,2],Xproj[:,0],alpha=0.01,c='g')
ax[1][2].scatter(Xproj[:,1],Xproj[:,0],alpha=0.01,c='g')
# this is scatter plot for the data after PCA
ax[0][0].set_ylabel(r'pairs of features in data')
ax[1][0].set_ylabel(r'pairs of principal components')
```

Apply PCA on a small face dataset in order to familiarise yourself with how "eigenfaces" are generated. This was a popular application context. The procedure is much the same as above, with the only additional steps being the resizing of 2-d images into 1-d arrays and back.

```python
dat=spio.loadmat("./Imgs.mat")
```

```python
images=dat['Imgs']
images.shape
imat=images.T
print(np.shape(imat))
```

There are 84 images of size (57,72) pixels each stored as a vector of length 4104. imat is a matrix with 84 rows (N) and 4104 columns (p).

```python
fig, ax = plt.subplots(nrows=12,ncols=7,figsize=(17,18))
for i in range(12):
    for j in range(7):
        ax[i][j].imshow(np.reshape(imat[i*7+j,:],(57,72)),cmap="gray")
```

First we extract the singular vectors and singular values of the mean subtracted image data. Then we compute the fraction of the total variance captured by 1, 2, ..., 84 principal components.

```python
meanim=np.mean(imat,axis=0)
devim=imat-meanim
[u, s, v] = np.linalg.svd(devim)
frac_variance = s**2/np.sum(s**2)
cumfrac_variance = np.zeros(84)
for i in range(84):
    cumfrac_variance[i] = np.sum(frac_variance[0:i+1])
```

## Eigenfaces
Now plot the mean face and a few eigenfaces. Projections along the principal components need to be performed.

```python
fig, ax = plt.subplots(nrows=2,ncols=4,figsize=(15,7))
ax[0][3].imshow(np.resize(meanim,(57,72)),cmap="gray")
ax[0][3].set_xlabel('mean',fontsize=18)
for i in range(2):
    for j in range(3):
        sid = 3*i+j
        ax[i][j].imshow(np.resize(np.dot(devim.T,u[sid]),(57,72)),cmap="gray")
        ax[i][j].set_xlabel(sid+1,fontsize=18)
ax[1][3].imshow(np.resize(np.dot(devim.T,u[29]),(57,72)),cmap="gray")
ax[1][3].set_xlabel(30,fontsize=18)
```

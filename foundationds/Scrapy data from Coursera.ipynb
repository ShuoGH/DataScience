### going to use the urlopen to open the website and save it into the bs4

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# if has Chinese, apply decode()
html = urlopen("https://www.coursera.org/learn/zika").read().decode('utf-8')
print(html)
# 这个可以抓取这个页面的element
```

```python
soup= BeautifulSoup(html, features='lxml')
```

**can use the `find()` to get the tag in the html page`**

```python
# img_name=soup.find("img",class_="rectangularXl_ltk00o")
# type(img_name)
# img_name.contents[0]

# get the university name 
uniname=soup.find("img",class_="rectangularXl_ltk00o")["title"]
```

```python
# get the course name 
cname=soup.find('h1',class_="H2_1pmnvep-o_O-weightNormal_s9jwp5-o_O-fontHeadline_1uu0gyz max-text-width-xl m-b-1s").contents[0]
```

```python
# because there maybe more than one instructors 
instructorList=soup.find_all('h3',class_="H2_1pmnvep-o_O-weightBold_uvlhiv-o_O-bold_1byw3y2 font-lg")
for i in instructorList:
    print(i.contents[0].contents[0])
    print("here you can jump into the new website for the lecture page")
# print(instructor)
```

**to jump into the website of the instructor**

    - to find the courses he teaches

```python
# get the category for the course 
cate=soup.find_all("div",class_="BreadcrumbItem_1pp1zxi")
for i,j in enumerate(cate):
    print(i,"\n",j.contents[0].contents[0],"\n")
#     can get the contents in the breadcrumbs
```

```python
# I find that how to get the language of the course. 
# although the order of these items are different sometimes. but 
# the language always be the last one. so, i can get the last one to get the language
# get the language of the course 
# there are some problems in this part.
lan=soup.find_all("h4",class_="H4_1k76nzj-o_O-weightBold_uvlhiv-o_O-bold_1byw3y2 m-b-0")[-1]
for i in lan:
    print(i)
# after get the language, can get the subtitles of this courses
```

```python
# get the captions?
lan=soup.find_all("div",class_="Box_120drhm-o_O-displayflex_poyjc m-b-2 p-t-1s")[-1].contents[1]
for i in lan:
    print(i,"\n")
#     now I can get the subtitles of the courses
```

```python
# get the rates
rates=soup.find("span",class_="H4_1k76nzj-o_O-weightBold_uvlhiv-o_O-bold_1byw3y2 m-l-1s m-r-1 m-b-0").contents[0]
print(rates)
```

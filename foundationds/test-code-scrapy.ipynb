### going to use the urlopen to open the website and save it into the bs4

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# if has Chinese, apply decode()
html = urlopen("https://www.coursera.org/learn/machine-learning").read().decode('utf-8')
print(html)
# can use beautifulsoup to scrape the content of website 
```

```python
soup= BeautifulSoup(html, features='lxml')
```

```python
# i use the file I/O to write the result to the text file 
codecs.open("soup-html.txt","w",'utf-8').write(str(soup))
# type(soup)
```

**can use the `find()` to get the tag in the html page`**

```python
# get the course name 
cname=soup.find('h1',class_="H2_1pmnvep-o_O-weightNormal_s9jwp5-o_O-fontHeadline_1uu0gyz max-text-width-xl m-b-1s").contents[0]
cname
```

```python
# get the category for the course 
cate=soup.find_all("div",class_="BreadcrumbItem_1pp1zxi")
for i,j in enumerate(cate):
    print(i,"\n",j.contents[0].contents[0],"\n")
#     can get the contents in the breadcrumbs
```

```python
# get the instructors list and his href. then get his url 
instructorList=soup.find_all('h3',class_="H2_1pmnvep-o_O-weightBold_uvlhiv-o_O-bold_1byw3y2 font-lg")
for i in instructorList:
    print(i.contents[0])
    print(i.contents[0]["href"])
#     print("here you can jump into the new website for the lecture page")
# print(instructor)
```
